# Two normal distributions with same mean and different standard deviations

```python
import numpy as np
import seaborn as sns

a = np.random.normal(60, 10, 100000)
sns.distplot(a, hist=True, kde=True, bins=40)

b = np.random.normal(60, 20, 100000)
sns.distplot(b, hist=True, kde=True, bins=40)
```

about 68% of the values in a are in [60,70] range

about 68% of the values in b are in [40,80] range

```python
import scipy.stats as st
st.norm.cdf(70, loc = 60, scale = 10) - st.norm.cdf(50, loc = 60, scale = 10)
```

```python
print(st.norm.cdf(80, loc = 60, scale = 20) - st.norm.cdf(40, loc = 60, scale = 20))
```

```python
# count how many of data sample would be less than arbitrary value
# count how many of a less than r
# sum(a <= r)/len(a) becomes percentage

plt.hist(a, bins = 20, density = True)
plt.hist(b, bins = 20, density = True, alpha = .5)
```

```python
def calculate_cdf(x, data):
    return np.sum(x < data)

def calculate_cdf_percentage(x, data):
    return np.sum(x < data)/len(x)

```

```python
# calculate the CDF of a. for r in range(np.mean(a))
cdf_a = [calculate_cdf(a, r)/len(a) for r in range(int(np.min(a)), int(np.max(a)))]

cdf_b = [calculate_cdf(b, r)/len(b) for r in range(int(np.min(b)), int(np.max(b)))]
```

```python
plt.plot(range(int(np.min(a)), int(np.max(a))), cdf_a)
plt.plot(range(int(np.min(b)), int(np.max(b))), cdf_b)
# cdf says 50% of our data(for a and b) is less than 60 (our mean)
```
